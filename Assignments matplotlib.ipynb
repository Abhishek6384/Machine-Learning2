```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# Exercise 1: Read Total profit of all months and show it using a line plot
df=pd.read_csv('company_sales_dataset.csv')
df
```

```python
plt.plot(df['month_number'],df['total_profit'],label='profit')
plt.title('line plot')
plt.xlabel('month_number')
plt.ylabel('total_profit')
plt.legend()

plt.show()
```

```python
# Exercise 2: Get total profit of all months and show line plot with the following Style properties

plt.plot(df['month_number'],df['total_profit'],marker='o',color='r',label='Sold units number',linewidth=3,linestyle='--')
plt.title('line plot')
plt.xlabel(' Month Number')
plt.ylabel('Sold units number')
plt.legend(loc='lower right',shadow=True,fontsize='large')

plt.show()
```

```python
# Exercise 3: Read all product sales data and show it  using a multiline plot
plt.figure(figsize=(10,7))
plt.plot(df['month_number'],df['facecream'],label='facecream')
plt.plot(df['month_number'],df['facewash'],label='facewash')
plt.plot(df['month_number'],df['toothpaste'],label='toothpaste')
plt.plot(df['month_number'],df['bathingsoap'],label='bathingsoap')
plt.plot(df['month_number'],df['shampoo'],label='shampoo')
plt.plot(df['month_number'],df['moisturizer'],label='moisturizer')
plt.xlabel(' Month Number')
plt.ylabel('All product sales')
plt.legend(loc='upper left')
plt.show()
```

```python
# Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot
plt.scatter(df['month_number'],df['toothpaste'],label='toothpaste',color='g')
plt.legend(loc='upper left')
plt.xlabel(' Month Number')
plt.ylabel('units')
plt.title('Excecise 4')
plt.show()
```

```python
# Exercise 5: Read face cream and facewash product sales data and show it using the bar chart
xpos=np.array(df['month_number'])
plt.figure(figsize=(10,7))
plt.bar(xpos-0.2,df['facecream'], width=0.4, label="facecream")
plt.bar(xpos+0.2,df['facewash'], width=0.4,label="facewash")

plt.xticks(df['month_number'])
plt.ylabel("units")
plt.title('facecream VS facewash')
plt.legend()
plt.show()
```

```python

# Exercise 7: Read the total profit of each month and show it using the histogram to see the most common profit ranges
plt.figure(figsize=(10,7))
plt.hist(df['total_profit'],rwidth=0.5,color='g',label='total_profit')
plt.xlabel('total_profit')
plt.ylabel('frequency')
plt.legend()
plt.show()
```

```python
# Exercise 9: Read Bathing soap facewash of all months and display it using the Subplot
plt.figure(figsize=(10,7))


plt.subplot(2,1,1)
plt.bar(df['month_number'],df['bathingsoap'], width=0.4, label="bathingsoapp",color='r')
plt.xlabel('bathingsoap')
plt.ylabel('month number')
plt.title('bathingsoap')
plt.legend()
plt.show()

```

```python
# Exercise 10: Read all product sales data and show it using the stack plot


lables=['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']


plt.figure(figsize=(10,7))
plt.stackplot(df['month_number'],df['facecream'],df['facewash'],df['toothpaste'],df['bathingsoap'],df['shampoo'],df['moisturizer'],labels=lables)
plt.xlabel('all products')
plt.ylabel('total units')
plt.legend(loc='upper left')
plt.show()
```
